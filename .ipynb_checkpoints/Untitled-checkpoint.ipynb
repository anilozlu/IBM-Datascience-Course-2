# Recommendation System for Spotify Users
The method we are going to employ is called User-Based Collaborative Filtering. 

## How it works:
We are only concerned with the past actions of users. (As the data we have only has playlists within, we will treat each playlist as a seperate user.)

Once we have the data for past user interactions, we can construct a user-item interaction matrix.

(Image source: https://buomsoo-kim.github.io/recommender%20systems/2020/09/25/Recommender-systems-collab-filtering-12.md/)

Once we construct this matrix, we can choose any user row within the matrix and use it to represent the user we want to make recommendations to. We simply calculate cosine similarities between that user and every other one. Then, we recommend items based on the actions of other similar users.

To sum the algorithm up, we match similar users using their past interactions, then recommend users items based on the actions of similar profiles.

```python
import pandas as pd
import numpy as np
from tqdm.auto import tqdm
from itertools import chain
import json
import os
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
```

```python
def prepare_data(data_folder_path):
    ## Spotify dataset comes in pieces, so we need to get all
    ## the filenames in the dataset.
    json_list = os.listdir(data_folder_path)
    ## iterate over the dataset files and get unique track_uris for
    ## each track in each playlist
    data = []
    track_information = []
    for json_file in tqdm(json_list[:10]):
        json_path = os.path.join("data", json_file)
        with open(json_path, "r") as f:
            data_part = json.load(f)
        ## get "playlists" from json
        playlists = data_part["playlists"]
        ## get only the tracks, not the info
        playlists = [p["tracks"] for p in playlists]
        ## get each track_uri in each playlist, seperate tracks with a 
        ## space
        playlist_tracks = [
            " ".join(t["track_uri"].split(":")[-1] for t in p)\
             for p in playlists
        ]
        for playlist in playlists:
            for track in playlist:
                track_information.append(
                    {
                        "name": track["track_name"],
                        "artist": track["artist_name"],
                        "track_uri": track["track_uri"].split(":")[-1]
                    }
                )
        data.append(playlist_tracks)
    track_information = pd.DataFrame(track_information)
    track_information = track_information.set_index("track_uri", drop=True)
    track_information = track_information[~track_information.index.duplicated()]
    ## get the content of each file in the same list
    spotify_playlists = []
    for playlist in data:
        spotify_playlists.extend(playlist)
    return spotify_playlists, track_information
```

```python
def train_transform(playlists):
    ## use a countvectorizer to transform every playlist
    ## into a vector
    cv = CountVectorizer(lowercase=False)
    playlist_vectors = cv.fit_transform(playlists)
    return cv, playlist_vectors
```

```python
def user_to_vec(cv, user):
    ## transform user into a vector using the same
    ## vectorizer model
    user_vector = cv.transform(user)
    return user_vector
```

```python
def get_similarities(user_vector, playlist_vectors):
    ## calculate every similarity between the user and the playlists
    ## using cosine similarity
    similarities = cosine_similarity(user_vector, playlist_vectors)
    ## put the similarities in a DataFrame. sort by similarity scores
    ## filter out if we have a similarity of 1, as it does not give
    ## us any recommendations.
    similarities = pd.DataFrame(similarities).T.rename(columns={0: "similarity"})
    similarities = similarities[similarities.similarity < 1-1e-5].sort_values("similarity")
    return similarities
```

```python
def get_recommendations(user_vector, similarities, playlists, track_information):
    ## get the top 5 similar playlists
    playlists_to_recommend_from = similarities.index[:5].tolist()
    ## get possible tracks to recommend
    tracks_to_recommend_from = []
    for i in playlists_to_recommend_from:
        similar_playlist = playlists[i].split(" ")
        similar_playlist = [t for t in similar_playlist if t not in user[0].split(" ")]
        tracks_to_recommend_from.append(similar_playlist)
    ## get 3 random tracks from each playlist and prepare them
    ## as recommendations
    recommendations = []
    for p in tracks_to_recommend_from:
        recommendations.extend(list(np.random.choice(p, replace=False, size=3)))
    recommendations_with_data = []
    for r in recommendations:
        recommendations_with_data.append(track_information.loc[r].to_dict())
    return recommendations_with_data
```

```python
spotify_playlists, track_information = prepare_data("./data")
cv, playlist_vectors = train_transform(spotify_playlists)
```

## Some Data Visualisation
Exploring user actions, by plotting the distribution of number of songs contained in playlists.

```python
playlist_counts = {idx:0 for idx in track_information.index}
for playlist in spotify_playlists:
    playlist = playlist.split(" ")
    for track in playlist:
        playlist_counts[track] += 1
playlist_counts_df = pd.DataFrame(index = playlist_counts.keys(), data = playlist_counts.values())
playlist_counts_df.index.name = "track_name"
playlist_counts_df = playlist_counts_df.rename(columns={0: "count"})
counted_data = {}
for i, group in playlist_counts_df.groupby("count"):
    counted_data[i] = group.shape[0]
```

```python
bins = np.arange(0, max(counted_data.keys()), 15)
_ = plt.figure(figsize=(12, 8))
_ = plt.hist(counted_data.keys(), bins[:-1], bins,
             weights=counted_data.values(), log=True)
_ = plt.xticks(ticks = _[1], labels = bins[:-1], rotation=45)
_ = plt.grid()
_ = plt.title("Playlists by Length")
_ = plt.xlabel("Playlist Length (Number of Songs in a Playlist)")
_ = plt.ylabel("Number of Playlists")
```
