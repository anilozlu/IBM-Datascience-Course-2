### This code will create a random normal distribution and plot it.

````python
import numpy as np
import matplotlib.pyplot as plt

dist = np.random.normal(size=(50000, 1))
_ = plt.hist(dist, bins=100)
_ = plt.grid()
````

A normal distribution:
* Is also known as Gaussian distribution
* Is a type of continuous probability distribution
* Based on a real random variable.
***
